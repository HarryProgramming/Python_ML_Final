```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the dataset

startup_big_data = pd.read_csv('./big_raw.csv', encoding='latin-1')
startup_big_data.head()

# make a new copy 
cleaned_big_data = startup_big_data.copy()
```

```python
# Select the columns that we want to keep

columns_to_keep = ['name', 'state_code', 'city', 'market', 'founded_at', 'founded_month',
                   'founded_quarter', 'funding_rounds', 'first_funding_at', 'last_funding_at',
                   'funding_total_usd', 'venture', 'angel', 'round_A', 'round_B', 'round_C', 'round_D',
                   'status', 'country_code']

# Drop the columns not selected
cleaned_big_data = cleaned_big_data.drop(columns=[col for col in cleaned_big_data.columns if col not in columns_to_keep])

cleaned_big_data.head()
```

```python
# Select the startups that are in USA

cleaned_big_data = cleaned_big_data[cleaned_big_data['country_code'] == 'USA']

# Reset the index of the dataframe
cleaned_big_data.reset_index(drop=True, inplace=True)

# Drop the 'country_code' column
cleaned_big_data = cleaned_big_data.drop('country_code', axis=1)

cleaned_big_data.head()
```

```python
# Rename the columns

cleaned_big_data = cleaned_big_data.rename(columns={
    'name': 'Startup_name',
    'state_code': 'State_code',
    'city': 'City',
    'market': 'Category',
    'founded_at': 'Founded_at',
    'founded_month': 'Founded_month',
    'founded_quarter': 'Founded_quarter',
    'funding_rounds': 'Funding_rounds',
    'first_funding_at': 'first_funding_at',
    'last_funding_at': 'last_funding_at',
    'funding_total_usd': 'Funding_total_usd',
    'venture': 'Has_VC',
    'angel': 'Has_angel',
    'round_A': 'Has_roundA',
    'round_B': 'Has_roundB',
    'round_C': 'Has_roundC',
    'round_D': 'Has_roundD',
    'status': 'Status'
})
```

```python
# Delete NA values

cleaned_big_data = cleaned_big_data.dropna()

# Delete rows with ' -   ' in the 'Funding_total_usd' column
cleaned_big_data = cleaned_big_data[cleaned_big_data['Funding_total_usd'] != ' -   ']
```

```python
# Add two ages columns

# Calculate and add the ages columns
cleaned_big_data['Age_first_funding_year'] = cleaned_big_data['first_funding_at'].str[:4].astype(int) - cleaned_big_data['Founded_at'].str[:4].astype(int)
cleaned_big_data['Age_last_funding_year'] = cleaned_big_data['last_funding_at'].str[:4].astype(int) - cleaned_big_data['Founded_at'].str[:4].astype(int)

# Delete the specified columns from the dataframe
columns_to_delete = ['first_funding_at', 'last_funding_at']
cleaned_big_data = cleaned_big_data.drop(columns_to_delete, axis=1)

cleaned_big_data.head()
```

```python
# Select the city we want to analyze

# Specify the cities to filter
cities_to_filter = ['San Francisco', 'New York', 'Mountain View', 'Palo Alto', 'Santa Clara', 'Austin',
                    'Seattle', 'San Mateo', 'Sunnyvale', 'San Jose', 'Cambridge', 'Menlo Park',
                    'Los Angeles', 'San Diego']

# Filter the dataframe for the specified cities
cleaned_big_data = cleaned_big_data[cleaned_big_data['City'].isin(cities_to_filter)]

# Print the number of rows for each city with aligned counts
city_counts = cleaned_big_data['City'].value_counts()
max_city_length = max(city_counts.index.map(len))

for city, count in city_counts.items():
    print(f"{city:{max_city_length}}  {count}")

# Reset the index
cleaned_big_data = cleaned_big_data.reset_index(drop=True)

print(cleaned_big_data.shape)
cleaned_big_data.head()
```

```python
# Unify 'Category'

# Define the categories and corresponding keywords
category_keywords = {
    'web': ['Network', 'Web'],
    'games': ['Games'],
    'hardware&software': ['Hardware', 'Software', 'Messaging'],
    'biotechnology': ['Biotechnology'],
    'analytics': ['Analytics'],
    'e-commerce': ['E-commerce'],
    'mobile': ['Mobile'],
    'advertising': ['Advertising'],
    'semiconductors': ['Semiconductors'],
    'security': ['Security'],
    'clean_technology': ['Clean_technology'],
    'public_relations': ['Public_relations'],
    'social_media': ['Social', 'News', 'Video'],
    'fashion': ['Fashion'],
    'travel': ['Travel'],
    'art&music': ['Art', 'Music'],
    'health&medical': ['Health', 'Medical'],
    'finance': ['Finance', 'Real Estate']
}

# Replace the 'Category' column values based on the renaming logic
replaced_categories = set()

for category, keywords in category_keywords.items():
    mask = cleaned_big_data['Category'].str.contains('|'.join(keywords), case=False, na=False)
    cleaned_big_data.loc[mask, 'Category'] = category
    replaced_categories.add(category)

# Filter out rows that were not replaced
cleaned_big_data = cleaned_big_data[cleaned_big_data['Category'].isin(replaced_categories)]

print(cleaned_big_data.shape)

cleaned_big_data.head()
```

```python
# Unify 'Status'

# Define the condition for updating 'Status' column
condition = cleaned_big_data['Status'] == 'closed'

# Update 'Status' column based on the condition
cleaned_big_data['Status'] = np.where(condition, 0, 1)

cleaned_big_data.head()
```

```python
# Unify 'Founded_month' 

# Create a new column 'Founded_datetime' with datetime format
cleaned_big_data['Founded_datetime'] = pd.to_datetime(cleaned_big_data['Founded_at'])

# Extract the month from 'Founded_datetime' and update 'Founded_month' column
cleaned_big_data['Founded_month'] = pd.to_datetime(cleaned_big_data['Founded_datetime']).dt.month

# Delete the 'Founded_datetime' column
cleaned_big_data = cleaned_big_data.drop('Founded_datetime', axis=1)

cleaned_big_data.head()
```

```python
# Unify 'Has_VC', 'Has_angel', 'Round_A', 'Round_B', 'Round_C', 'Round_D'

# Update 'Has_VC' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_VC'] != 0, 'Has_VC'] = 1

# Update 'Has_angel' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_angel'] != 0, 'Has_angel'] = 1

# Update 'Round_A' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_roundA'] != 0, 'Has_roundA'] = 1

# Update 'Round_B' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_roundB'] != 0, 'Has_roundB'] = 1

# Update 'Round_C' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_roundC'] != 0, 'Has_roundC'] = 1

# Update 'Round_D' column where it is not equal to 0
cleaned_big_data.loc[cleaned_big_data['Has_roundD'] != 0, 'Has_roundD'] = 1

cleaned_big_data.head()
```

```python
# Filter out rows where 'Age_first_funding_year' and 'Age_last_funding_year' are smaller than 0

cleaned_big_data = cleaned_big_data[(cleaned_big_data['Age_first_funding_year'] >= 0) & (cleaned_big_data['Age_last_funding_year'] >= 0)]

cleaned_big_data.head()
```

```python
# Export the new dataset

cleaned_big_data.to_csv('big.csv', index=False)
```
